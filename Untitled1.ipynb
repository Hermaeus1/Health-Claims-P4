```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
import matplotlib as MP


#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("/insurance_data.csv")
application_df.head()
```

```python
# Remove rows with any null values
cleaned_df = application_df.dropna(subset=['age', 'region'])
```

```python
# Check for null values in the DataFrame and get columns with null values
columns_with_null = cleaned_df.columns[cleaned_df.isnull().any()]

# Print the specific columns with null values
if not columns_with_null.empty:
    print("Columns with null values:")
    for column in columns_with_null:
        null_count = cleaned_df[column].isnull().sum()
        print(f"{column}: {null_count} null values")
else:
    print("No columns with null values in the DataFrame.")
```
