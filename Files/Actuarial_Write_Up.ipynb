# Assumptions For Renewal Modeling
### Subscriber Death
- Patient terming is solely based on death
- Patient death is dependent on the Social Security Administration (SSA) Actuarial Life Table
### Subcriber Smoking Cessation
- Quitting rate is assumed to be the worst case scenario of 50% of smokers try to quit yearly and 3 in 5 of them are able to quit for a total probability of 0.5 * 0.6 = 0.3 (from CDC)
- If someone is not a smoker, they will not begin smoking frequently (assumed to have adequate EAP programming)
### Children Modeling
- Female subscribers are able to have children over the year, maximum of 1 child added
- Birth rates are based on Birth Rate of Women by age group

For the pricing aspect, future claims are simulated nia MCMC Monte Carlo Markov Chains with Contiunance Tables with 100 trials
-

For each trial all claims are summed, choosing a percentile of the data, then divided by the number of subscribers, then the number of months. An annual claims growth trend of 7% is assumed for the method and will be applied after simulation. Lastly, incurred but not reported reserve will not be included in the pricing pf premiums as no information on claims lag is provided in the data set.
