```python
import pandas as pd
import requests

url = 'https://www.ssa.gov/oact/STATS/table4c6.html'
html = requests.get(url).content
df_list = pd.read_html(html,skiprows=0,header=1)[0]
df = pd.DataFrame(df_list)
df.head()
```

```python
# renaming columns to account for gender, then dropping lives and expectancy
df.columns = ['Exact_Age','Male_Death_Prob','Male_Num_Lives','Male_Life_Expectancy','Female_Death_Prob','Female_Num_Lives','Female_Life_Expectancy']
df.head()
```

```python
final_life_df = df.drop(['Male_Num_Lives','Male_Life_Expectancy','Female_Num_Lives','Female_Life_Expectancy'],axis=1)
final_life_df.set_index('Exact_Age',inplace=True)
final_life_df.head(10)
```
